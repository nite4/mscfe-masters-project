# Read Data from Files

```python
print('*****Reading equity data*****')
df_equity = read_excel_sheets('data.xlsx')
display(df_equity.head())
list_equity = df_equity.Symbol.unique().tolist()

print('*****Reading cryptocurrency data*****')
df_crypto = read_excel_sheets('data_crypto.xlsx')
display(df_crypto.head())
list_crypto = df_crypto.Symbol.unique().tolist()

print('*****Concatenating both dfs*****')
df_full = pd.concat([df_equity, df_crypto])
display(df_full.sample(5))

print('*****Check for missing values*****\n', df_full.isna().sum())
```

```python
symbols = df_full['Symbol'].unique()
nrows, ncols = 9, 3  # Grid dimensions
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(12, 20))
fig.suptitle('5-minute Prices of the Potential Pair Trading Portfolio Components\n\n', fontsize=16)

axes = axes.flatten()
for idx, s in enumerate(symbols):
    ax = axes[idx]
    subset = df_full[df_full['Symbol'] == s]
    ax.plot(subset['OpenTime'], subset['Close'], label=s, color='dodgerblue')
    ax.set_title(f'Price of {s}  Training Set', fontsize=10)
    ax.set_xlabel('Date', fontsize=8)
    ax.set_ylabel('Price', fontsize=8)
    ax.tick_params(axis='both', labelsize=8)
    ax.legend(fontsize=6)
    if (len(s)>6) and ('ETF' not in s):
        ax.xaxis.set_major_locator(plt.IndexLocator(base=1500, offset=0))
    else:
        ax.xaxis.set_major_locator(plt.IndexLocator(base=250, offset=0))
    plt.setp(ax.xaxis.get_majorticklabels(), rotation=45, ha='right')

# Remove any unused subplots
for idx in range(len(symbols), len(axes)):
    fig.delaxes(axes[idx])

plt.tight_layout()
plt.savefig('output/portfolio_component_prices.png')
plt.show();
```
